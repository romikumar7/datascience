```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
plt.show()
```

```python
plt.figure()
# create a plot figure
# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))
# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)
# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
# Line plots
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
plt.plot(x, np.cos(x));
```

```python
# Adjusting colors
plt.plot(x,np.sin(x-0),color='blue') # specify color name
plt.plot(x,np.sin(x-1),color='g') # short color code (rgbcmyk)
plt.plot(x,np.sin(x-2),color='0.75') # grayscale between 0 and 1
plt.plot(x,np.sin(x-3),color='#FFDD44') # hex code (RRGGBB from 00 to FF)
plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3)) # RGB tuple, values from 0 to 1
plt.plot(x,np.sin(x-5),color='chartreuse'); # all html color name supported
```

```python
# adjusting linestyle
plt.plot(x,x+1,linestyle='solid')
plt.plot(x,x+2,linestyle='dashed')
plt.plot(x,x+3,linestyle='dashdot')
plt.plot(x,x+4,linestyle='dotted')
# above can also be done using codes for line styles
plt.plot(x,x+5,linestyle='-') # solid
plt.plot(x,x+6,linestyle='--') # dashed
plt.plot(x,x+7,linestyle='-.') # dashdot
plt.plot(x,x+8,linestyle=':') # dotted
```

```python
# both linestyle and colorcode can be combined too without using keyword
plt.plot(x,x+5,'-g') # solid green
plt.plot(x,x+6,'--c') # dashed cyan
plt.plot(x,x+7,'-.k') # dashdot black
plt.plot(x,x+8,':r') # dotted red
```

```python
# adjusting the plot : axis limit
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
# you can inverse the axis limits too
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

```python
# axis limits can be specified via axis function also
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5, 1.5])
# XMIN, XMAX, YMIN, YMAX
```

```python
# labeling plot
plt.plot(x, np.sin(x))
plt.title("A sin curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.title('A sin/cos curve')
plt.legend();
```

The primary difference of plt.scatter from plt.plot is that it can be used to create scatter plots where the properties of each individual point (size, face color, edge color, etc.) can be individually controlled or mapped to data.

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis') 
# alpha adjust the transparancy level value between 0 and 1 (0 means white no color), however 1 ( doesnot mean black)
plt.colorbar(); # show color scale
```

```python
# Visualizing 3D function using Density or Contour plots
plt.style.use('seaborn-white')
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
# meshgrid function is used to create a rectangular grid out of two given one-dimensional arrays representing the Cartesian indexing or Matrix indexing.
Z = f(X, Y)
plt.contour(X, Y, Z, colors='black');
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',cmap='RdGy')
plt.colorbar()
plt.axis('image');
```

* plt.imshow() doesnt accept an x and y grid, so you must manually specify the extent [xmin, xmax, ymin, ymax] of the image on the plot.
* plt.imshow() by default follows the standard image array definition where the origin is in the upper left, not in the lower left as in most contour plots. This must be changed when showing gridded data.
* plt.imshow() will automatically adjust the axis aspect ratio to match the input data; you can change this by setting, for example, plt.axis(aspect='image') to make x and y units match.

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
cmap='RdGy', alpha=0.5)
plt.colorbar();
```

```python
# Histograms, Binning and Density

plt.style.use('seaborn-white')
data = np.random.randn(1000)
plt.hist(data);
```

```python
plt.hist(data, bins=30, density=True, alpha=0.5, histtype='stepfilled', color='steelblue', edgecolor='none');
```

```python
# 2d histogram

mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
# creates bins using squares 
```

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
# creates bins using hexagonal
```

##### choosing the colormap
You should be aware of three different categories of colormaps:
* Sequential colormaps - These consist of one continuous sequence of colors (e.g., binary or viridis ).
* Divergent colormaps - These usually contain two distinct colors, which show positive and negative deviations from a mean (e.g., RdBu or PuOr ).
* Qualitative colormaps - These mix colors with no particular sequence (e.g., rainbow or jet ).
